```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.metrics import accuracy_score,classification_report,mean_squared_error
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("SugarPrediction.csv")
df
```

```python
df.drop(columns="PPBS",inplace=True)
```

```python
train_gender = {'M': 1,'F':0}
df.Gender =[train_gender[item] for item in df.Gender]
```

```python
train_genetic = {'Yes': 1,'No':0,'Not Sure':2}
df.Genetic =[train_genetic[item] for item in df.Genetic]
```

```python
train_lifestyle = {'Normal': 1,'Active':0,'Sedentary':2}
df.Lifestyle =[train_lifestyle[item] for item in df.Lifestyle]
```

```python
train_Area = {'Urban': 1,'Metro':0,'Rural':2}
df.Area =[train_Area[item] for item in df.Area]
```

```python
train_Stress = {'Normal': 1,'Low':0,'High':2}
df.Stress =[train_Stress[item] for item in df.Stress]
```

```python
train_Diabetic = {'Border': 1,'No':0,'Yes':2}
df.Diabetic =[train_Diabetic[item] for item in df.Diabetic]
df
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

X = df.drop(columns =['FBS','Diabetic'], axis=1)
Y = df['FBS']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=101)

```

```python
reg.fit(X_train,y_train)
```

```python
preds = reg.predict(X_test)
preds
```

```python
preds = np.floor(preds)
```

```python

print('RMSE:', np.sqrt(mean_squared_error(y_test, preds)))
```
