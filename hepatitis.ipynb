```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import accuracy_score,classification_report
```

```python
df = pd.read_csv("Training.csv")
data = df.loc[df['prognosis'].isin(["hepatitis A"])]
data1 = df.loc[df['prognosis'].isin(["Hepatitis B"])]
data2 = df.loc[df['prognosis'].isin(["Hepatitis C"])]
df = pd.concat([data,data1,data2], axis=0, join='inner')
df
```

```python
df.drop(columns='Unnamed: 133',inplace=True)
df
```

```python
A = df[["prognosis"]] # diseases 
B = df.drop(["prognosis"],axis=1) # symptoms 
x_train, x_test, y_train, y_test = train_test_split(B,A,test_size=0.2) # 20:80 
```

```python
from sklearn import datasets
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
chi2_selector = SelectKBest(chi2,k=132)
X_kbest = chi2_selector.fit_transform(B, A)
```

```python
p_values=pd.Series(X_kbest[0])
pd.set_option('display.max_columns', None)

p_values.index=B.columns
x1 = p_values.sort_values(ascending=False)
x1
```

```python
applicable = ["loss_of_appetite","vomiting","diarrhoea","mild_fever","nausea","yellowish_skin","dark_urine","abdominal_pain","fatigue"]
x1_train = x_train[applicable]
x1_test = x_test[applicable]
```

```python
from sklearn.ensemble import RandomForestClassifier


mod = RandomForestClassifier(n_estimators = 100,n_jobs = 5, criterion= 'entropy',random_state = 42)
mod = mod.fit(x1_train,y_train.values.ravel())
pred = mod.predict(x1_test)
```

```python
print(accuracy_score(y_test, pred))
```

```python
report = classification_report(y_test, pred, output_dict=True)
pd.DataFrame(report).transpose()
```
